```python
import glob
import os
import re
from nltk.tokenize import sent_tokenize, word_tokenize

file_folder = "data/*.txt"   # sesuaikan dengan folder kamu


idx = 0
dict_global = {}           # <‚Äî ini yang penting!
files_with_index = {}
Stopwords = set(['first', 'activity', 'defense', 'events', 'anne', 'libel', 'four', 'university', 'attorney', 'served'])  # contoh stopwords

def remove_special_characters(text):
    return re.sub('[^a-zA-Z\s]', '', text)

def finding_all_unique_words_and_freq(words):
    freq = {}
    for w in words:
        freq[w] = freq.get(w, 0) + 1
    return freq

# üîÅ Loop membaca semua file teks
for file in glob.glob(file_folder):
    print(file)
    fname = file
    with open(file, "r", encoding="utf-8", errors="ignore") as f:
        text = f.read()

    text = remove_special_characters(text)
    text = re.sub(r'\d', '', text)
    sentences = sent_tokenize(text)
    words = word_tokenize(text)
    words = [word for word in words if len(word) > 1]
    words = [word.lower() for word in words]
    words = [word for word in words if word not in Stopwords]

    dict_global.update(finding_all_unique_words_and_freq(words))
    files_with_index[idx] = os.path.basename(fname)
    idx += 1

print("Selesai ‚úÖ")
print("Total kata unik:", len(dict_global))
```

```python
# --- Struktur data Node dan Linked List ---
class Node:
    def __init__(self, doc_id=None, freq=None):
        self.doc_id = doc_id
        self.freq = freq
        self.nextval = None

class SlinkedList:
    def __init__(self):
        self.head = None


# ==========================================
# PROSES PEMBENTUKAN INDEX BOOLEAN MODEL
# ==========================================

linked_list_data = {}

# Pastikan variabel sudah ada
if 'unique_words_all' not in locals():
    unique_words_all = set(dict_global.keys())

# Inisialisasi linked list kosong untuk setiap kata unik
for word in unique_words_all:
    linked_list_data[word] = SlinkedList()   # tidak perlu Node() kosong

word_freq_in_doc = {}
idx = 1

for file in glob.glob(file_folder):
    with open(file, "r", encoding="utf-8", errors="ignore") as f:
        text = f.read()

    # Bersihkan dan tokenisasi teks
    text = remove_special_characters(text)
    text = re.sub(r'\d', '', text)
    sentences = sent_tokenize(text)
    words = word_tokenize(text)
    words = [word for word in words if len(word) > 1]
    words = [word.lower() for word in words]
    words = [word for word in words if word not in Stopwords]

    # Hitung frekuensi kata per dokumen
    word_freq_in_doc = finding_all_unique_words_and_freq(words)

    # Tambahkan node ke linked list untuk setiap kata
    for word, freq in word_freq_in_doc.items():
        new_node = Node(idx, freq)

        # Jika linked list kata masih kosong ‚Üí jadikan head
        if linked_list_data[word].head is None:
            linked_list_data[word].head = new_node
        else:
            # Jika sudah ada node ‚Üí sisipkan di akhir
            current = linked_list_data[word].head
            while current.nextval is not None:
                current = current.nextval
            current.nextval = new_node

    idx += 1

print("‚úÖ Proses selesai!")
print("Jumlah kata unik:", len(linked_list_data))
print("Contoh kata:", list(linked_list_data.keys())[:10])
```

```python
# === PENCARIAN OTOMATIS UNTUK KATA "badminton" ===

query = "badminton"
print(f"üîç Query: {query}")

# Normalisasi dan tokenisasi
query_tokens = word_tokenize(query.lower())
different_words = query_tokens
connecting_words = []

print("Operator logika:", connecting_words)
print("Kata yang dicari:", different_words)

total_files = len(files_with_index)
zeroes_and_ones_of_all_words = []

# Representasi biner untuk kata 'badminton'
for word in different_words:
    if word in unique_words_all:
        zeroes_and_ones = [0] * total_files
        linkedlist = linked_list_data[word].head
        print(f"Memproses kata: {word}")
        while linkedlist is not None:
            zeroes_and_ones[linkedlist.doc_id - 1] = 1
            linkedlist = linkedlist.nextval
        zeroes_and_ones_of_all_words.append(zeroes_and_ones)
    else:
        print(f"‚ö†Ô∏è Kata '{word}' tidak ditemukan di indeks.")
        zeroes_and_ones_of_all_words.append([0] * total_files)

# Karena hanya satu kata, langsung ambil hasilnya
final_result = zeroes_and_ones_of_all_words[0]

# Tampilkan file yang cocok
files = [files_with_index[i + 1] for i, bit in enumerate(final_result) if bit == 1]

print("\nüìÇ Dokumen yang mengandung kata 'badminton':")
if files:
    for f in files:
        print("-", f)
else:
    print("Tidak ada dokumen yang mengandung kata tersebut.")
```
